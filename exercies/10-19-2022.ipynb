```python
x = np.linspace(1,10,20)
```

```python
y1 = 2 * x + 3  # a linear function
y2 = x**2 - 5 * x + 3  # a quadratic function
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x, y1,
    label='y = 2X + 3',
    linestyle = 'dashdot', 
    marker = 's',
    color = 'black')

ax.plot(x,y2,label='$y = x^2 - 5x + 3$', linestyle = 'dashed') #latex
ax.set_xlabel('population')
ax.set_ylabel('GDP')
ax.legend()
ax.set_title('two function'.title())
```
