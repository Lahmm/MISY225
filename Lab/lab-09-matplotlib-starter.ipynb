```python
# import matplotlib and finish the figure
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
x = np.linspace(-10,10,50)
y1 = 2*x
y2 = x**2 + 2*x + 1
y3 = x**3
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,15))

#subplot [0,0]
ax[0,0].plot(x,y1,
    label = 'y = 2x',
    marker = 'o',
    linestyle = 'dashdot'
)
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y')
ax[0,0].legend()
ax[0,0].set_title('Linear function')

#subplot [0,1]
ax[0,1].plot(x,y2,
    label = '$y = x^2 + 2x + 1$',
    marker = 'd',
    color = 'green',
    linestyle='dotted'
)
ax[0,1].set_xlabel('x')
ax[0,1].set_ylabel('y')
ax[0,1].legend()
ax[0,1].set_title('Quadratic function')

# subplot [1,0]
ax[1,0].plot(x,y3,
    label = '$y = x^3$',
    marker = 's',
    color = 'yellow',
    linestyle='solid'
)
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('y')
ax[1,0].legend()
ax[1,0].set_title('Power function')
```
